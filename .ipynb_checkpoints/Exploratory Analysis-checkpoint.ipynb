## chi-square

```python
import os
import pandas
#Changing the current working directory
chisquare = dataset
chisquare.rename(columns = {'Life expectancy ':'LifeExpectancy'}, inplace = True)
categorical_col = ['Developing','Developed']
print(chisquare)
```

```python
chisqt = pandas.crosstab(chisquare.Status, chisquare.LifeExpectancy, margins=True)
print(chisqt)
```

```python
chisqt.iloc[1][0:5]
```

```python
chisqt.iloc[1][0:5].values
```

```python
chisqt.iloc[1][0:20].values
```

```python
from scipy.stats import chi2_contingency 
value = np.array([chisqt.iloc[1][0:20].values,
                  chisqt.iloc[2][0:20].values])
print(chi2_contingency(value)[0:3])
```

## Chi square for array

```python
from scipy.stats import chi2_contingency 
 
info = [[100, 200, 300], [50, 60, 70]] 
print(info)
stat, p, dof, ex= chi2_contingency(info) 
 
print(dof)
print(stat)
print(p)
print(ex)
 
significance_level = 0.05
print("p value: " + str(p)) 
if p <= significance_level: 
    print('Reject NULL HYPOTHESIS') 
else: 
    print('ACCEPT NULL HYPOTHESIS') 
```

## Correlation

```python
import os
import pandas as pd
import numpy as np
import seaborn as sn
 
# Loading the dataset
corre = dataset
corre.rename(columns = {'Life-expectancy':'LifeExpectancy'}, inplace = True)
 
# Numeric columns of the dataset
numeric_col = ['LifeExpectancy','Population','Alcohol']
 
# Correlation Matrix formation
corr_matrix = corre.loc[:,numeric_col].corr()
print(corr_matrix)
 
#Using heatmap to visualize the correlation matrix
sn.heatmap(corr_matrix, annot=True)
```
